```python
from sys import path
import re
import pandas as pd
from bs4 import BeautifulSoup
from pathlib import Path
```

```python
file_path = Path('test2.html')

html = file_path.read_text(encoding='utf-8', errors='replace')
soup = BeautifulSoup(html)
```

```python
players = soup.find_all('article', class_='player')
len(players)
```

```python
p1 = players[0]
p2 = players[0+1]

p1.find('div', class_='player-info').find('div').find('strong').getText()
p1.find('div', class_='player-card').get('data-athlete-id')
p1.find('footer').find('strong').getText().split(' ', 1)

t1 = p1.find('div', class_='player-card player-loser')
t1 is None
```

```python
def parse_article_data(article):
    if article.find('footer').find('strong').getText() == '':
        return None
    player_name = article.find('div', class_='player-info').find('div').find('strong').getText()
    player_id = article.find('div', class_='player-card').get('data-athlete-id')
    kata_nr, kata_name = article.find('footer').find('strong').getText().split(' ', 1)
    winner = article.find('div', class_='player-card player-loser') is None

    return re.sub(r"\s+", " ", player_name).strip(), player_id, kata_nr, kata_name, winner
```

```python
print(parse_article_data(p1))
```

```python
for player in players:
    print(parse_article_data(player))
```
