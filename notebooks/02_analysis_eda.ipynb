```python
from pathlib import Path
import pandas as pd
```

```python
PARSED_DIR = Path("../data/parsed")

def read_all_parsed_csv(parsed_dir: Path = PARSED_DIR) -> pd.DataFrame:
    files = sorted(parsed_dir.glob("**/*.csv"))

    if not files:
        raise FileNotFoundError(f"No CSV files found in {parsed_dir}")

    dfs = []

    for file in files:
        df = pd.read_csv(file)

        relative_path = file.relative_to(parsed_dir)
        df["parsed_file"] = str(relative_path)

        if len(relative_path.parts) >= 2:
            df["tournament_id_from_path"] = relative_path.parts[0]
            df["category_from_path"] = file.stem

        dfs.append(df)

    return pd.concat(dfs, ignore_index=True)
```

```python
def read_single_parsed_csv(path: Path) -> pd.DataFrame:
    df = pd.read_csv(path)
    return df
```

```python
def summarize_kata_usage(df: pd.DataFrame) -> pd.DataFrame:
    summary = (
        df.groupby(["category", "kata_name"])
        .agg(
            performances=("kata_name", "size"),
            wins=("winner", "sum"),
            unique_athletes=("player_id", "nunique"),
        )
        .reset_index()
    )

    summary["total_category_performances"] = (
        summary.groupby("category")["performances"].transform("sum")
    )

    summary["usage_pct"] = (
        summary["performances"] / summary["total_category_performances"] * 100
    )

    summary["win_rate"] = summary["wins"] / summary["performances"] * 100

    summary = summary.sort_values(
        ["category", "performances", "win_rate"],
        ascending=[True, False, False],
    )

    return summary
```
